```python
def dkfunc(i) :
    back = "";
    if i % 15 == 0:
        back = "Fizz Buzz"
    elif i % 5 == 0 :
        back = "Buzz"
    elif i % 3 == 0 :
        back = "Fizz"
    else :
        back =""
    return print(back)
```

```python
dkfunc(15)
```
