```python
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
def count(ser) : 
    return ser.size
```

```python
count(tips['tip'])
```

```python
# 생성한 count 함수를 estimator에 적용
sns.barplot(x='sex', y='tip', data=tips, estimator=count)
```

```python
def over_2doll(ser) :
    return (tips['tip'] > 2).sum()
```

```python
sns.barplot(x='sex', y='tip', data=tips, estimator=over_2doll)
```

```python
sns.barplot(x='sex', y='tip', hue='time', data=tips, color='red')
```

```python
sns.barplot(x='sex', y='tip', hue='time', data=tips, palette=['red', 'cyan'])
```

```python
sns.barplot(x='sex', y='tip', hue='time', data=tips, palette='Blues_d')
```

```python
# 신뢰구간 변경 ci 옵션 설정
sns.barplot(x='sex', y='tip', hue='time', data=tips, ci=50)
```

```python
sns.barplot(x='sex', y='tip', hue='time', data=tips, capsize=0.2, errwidth=5, errcolor='red')
```
